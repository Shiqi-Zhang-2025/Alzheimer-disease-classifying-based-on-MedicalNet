```python
import csv
import matplotlib.pyplot as plt

# Load data from CSV files
with open('/validation/training_accuracy.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader, None)  # Skip the header row
    epochs_acc = [float(row[1]) for row in reader]

with open('/validation/training_loss.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader, None)  # Skip the header row
    epochs_loss = [float(row[1]) for row in reader]

# Plot training accuracy
epochs = range(1, len(epochs_acc) + 1)
plt.plot(epochs, epochs_acc, 'bo', label='Training accuracy')
plt.title('Training accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training loss
plt.plot(epochs, epochs_loss, 'b', label='Training loss')
plt.title('Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
